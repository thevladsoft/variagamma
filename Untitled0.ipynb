Comenzamos por ver cuantos archivos hay, cual es el más largo (para determinar el tamaño máximo de la matríz a usar) y leemos sus nombres.

```python
largo =!wc -l Fase*|sort -g|tail -n2|head -n1 | cut -d " " -f4
ancho=!ls Fase*|wc -l
print("existen "+ancho[0]+" archivos. El más largo tiene "+largo[0]+" lineas")
archivos=!ls Fase*
```

Cargamos la data completa. Como está toda a su manera (en archivos distintos), la extraeremos para llenar matrices.

Las columnas de los archivos FaseKroll vienen a ser:

-    Número de tiempo de vuelo distinto
-    Valor de Gamma
-    Valor de la Porosidad
-    Valor de la Fase de vuelo
-    Valor del tiempo de vuelo 
-    Valor del acumulado de las fases
-    1/Gamma


```python
datas=[]
for archie in archivos:
    datas.append(np.loadtxt(archie).T)
# for data in datas:
#     plt.plot(data[1],data[4])
#     plt.show()
# data[0]
# type(data),type(data[0])
```

Creamos las matrices a usar:

- X y Y son las coordenadas de los puntos.
    - Las matrices tiene largo el máximo numero de puntos que tiene un archivo (si no hubiese bifurcación ese sería el número valores de gamma diferentes).
    - El ancho es el número de archivos, o lo que es lo mismo, el numero de valores de phi.
- Zup, incluye los valores de tiempo de vuelo, pero en la zona de la bifurcación solo la rama superior.
- Zdown, solo incluye los valores de tiempo de vuelo para la rama inferior de la bifurcación.

Todos inicializados a *nan*, para que puedan quedar espacios vacios.

```python
X=ndarray((int(largo[0]),int(ancho[0])))*nan
Y=X.copy();Zup=X.copy();Zdown=X.copy();Zsalto=X.copy()
print("las matrices tienen dimensiones "+str(Zdown.shape[0])+" por "+str(Zdown.shape[1]))
# meshgrid([nan,nan],[nan,nan])
for i in range(len(datas)):
    salto=0
    for j in range(np.unique(datas[i][1]).shape[0]):
        X[j,i]=np.unique(datas[i][1])[j]
        Y[j,i]=datas[i][2,0]
        if sum(np.unique(datas[i][1])[j]==datas[i][1,:])==1 and salto:
            salto=2
            Zsalto[j,i]=datas[i][4,np.unique(datas[i][1])[j]==datas[i][1,:]]
        if salto<2:
            Zup[j,i]=datas[i][4,np.unique(datas[i][1])[j]==datas[i][1,:]].max()
        if sum(np.unique(datas[i][1])[j]==datas[i][1,:])>1:  
            salto=1
            Zdown[j,i]=datas[i][4,np.unique(datas[i][1])[j]==datas[i][1,:]].min()
        

plot(datas[i][1],datas[i][4],"o")
plt.title("un ejemplo")
plt.show()
```

```python
data4min=[]
data4max=[]
for data in datas:
    data4max.append(np.max(data[4]))
    data4min.append(np.min(data[4]))
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(9,6))
ax = fig.gca(projection='3d')
mapa=cm.OrRd_r
ax.plot_surface(X, Y,Zup,rstride=1, cstride=1,cmap=mapa,linewidth=0.0, antialiased=True,alpha=1.,
                vmin=max(data4min),vmax=max(data4max))
ax.view_init(elev=41,azim=-103)
plt.xlabel(r"$\Gamma$",fontsize=22)
plt.ylabel(r"$\phi$",fontsize=22)
ax.set_zlabel(r"$t_{vuelo}$",fontsize=18)
# ax.set_ylabel
ax.plot_surface(X, Y,Zdown,rstride=1, cstride=1,cmap=mapa,linewidth=0.02, antialiased=True,alpha=0.3,
                vmin=max(data4min),vmax=max(data4max))
ax.plot_surface(X, Y,Zsalto,rstride=1, cstride=1,cmap=mapa,linewidth=0, antialiased=True,alpha=1.,
                vmin=max(data4min),vmax=max(data4max))
# plt.title(mapa.name)
plt.show()
```
